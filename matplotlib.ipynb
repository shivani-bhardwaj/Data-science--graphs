```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi,np.pi,100)
```

```python
plt.plot(x,np.sin(x))
```

```python
theta= np.linspace(-np.pi,np.pi,100)  #generating a circle
radius = 10
```

```python
x=radius*np.sin(theta)
y=radius*np.cos(theta)    #equation=>x=rsin(t),y=rcos(t)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y)
plt.show()
```
